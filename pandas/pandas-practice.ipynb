```python
import pandas as pd
df=pd.read_csv("IRIS.csv")
df
```

```python
c1=df.copy()
c1 = c1[(c1["petal_length"] > 4.5) & (c1["species"] == "Iris-virginica")]
c1
```

```python
# 1. petal_ratio কলাম তৈরি করা
c1["petal_ratio"] = c1["petal_length"] / c1["petal_width"]



# 3. যদি species অনুযায়ী mean চাই
c1.groupby("species")["petal_ratio"].mean()
```
