```python
#melt && pivot
import pandas as pd

demo_pandas=pd.DataFrame({
    "Country":["USA","USA","India","India"],
    "year":[2020,2021,2020,2021],
    "sales":[100,120,90,110],
    "profit":[20,25,30,40],
    "tax":[10,20,30,40]
})
#print(demo_pandas)
melt_df=demo_pandas.melt(
    id_vars=["Country","year"],
    value_vars=["sales","profit"],
    var_name="metrices",
    value_name ="value"
    
)

# oppsite  melt to pivot
melt_df.pivot(
    index=["Country","year"],
    columns="metrices",
    values="value"
)
```
