```python
import json 
```

```python
def load_data(filename):
    with open (filename,"r")as f:
        data=json.load(f)
    return data

    
```

```python
load_data("store.json")
```

```python
def clean_data(data):
    text_to_num={"one":1,"two":2,"three":3,"four":4,"five":5}
    cleaned_data=[]
    unique_data=set()
    for user in data:
        temp_state=str(user['rating']).strip().lower()
        if temp_state in text_to_num:
          user['rating']=text_to_num[temp_state]
        missing_check=user.get("age")
        
        if missing_check==None:
            user["age"]=None
        if user['name'].strip() in unique_data:
            continue
        unique_data.add( user['name'])
        cleaned_data.append(user)
    return  cleaned_data
         
    
```

```python
clean_data(data)
```

```python
def average_rating(data):
    total_rating = 0
    count = 0

    for user in data:
        rating_num = float(user['rating'])
        total_rating += rating_num
        count += 1

    avg = total_rating / count
    return avg
```

```python
avg = average_rating(data)
print((avg))
```
