```python
#real data
#oscar  movie review
oscar_movies=[
    "the darak  night",
    "the hurt locker",
    "the kings a speech",
    "the artist",
    "Agro"
]
years=[2008,2009,2010,2011,2012]
oscar_revenew=[1005,170,427,133,232]
plt.plot(years,oscar_revenew)
plt.title("oscar movie  reveniew  in  each  year")
plt.xlabel("years")
plt.ylabel("Reveniew (in $M)")
```

```python
# multiple  dataset
non_oscar_movies=["slumdog Millionaire","Avatar","Inception","Hugo","Lincoln"]
non_oscar_revenew=[378,2788,185,275,100]
plt.plot(years,oscar_revenew, "o--r", label="oscar-winner")
plt.plot(years,non_oscar_revenew,"<-y", label="non_oscar_winner")
plt.title("oscar  vs  non  oscar  reviniew")
plt.xlabel("years")
plt.ylabel("revieniew")
plt.legend()
```

```python
# multiple  dataset  with  diffrent style
non_oscar_movies=["slumdog Millionaire","Avatar","Inception","Hugo","Lincoln"]
non_oscar_revenew=[378,2788,185,275,100]
plt.plot(years,oscar_revenew, color="red",marker="<",linestyle="-.",linewidth=5, label="oscar-winner")
plt.plot(years,non_oscar_revenew,"<-y", label="non_oscar_winner")
plt.title("oscar  vs  non  oscar  reviniew")
plt.xlabel("years")
plt.ylabel("revieniew")
plt.legend()
```
