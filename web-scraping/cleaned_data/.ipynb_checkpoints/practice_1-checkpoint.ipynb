```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

base_url = "https://quotes.toscrape.com"
url = "/page/1/"
ans = []

while url:
    r = requests.get(base_url + url)
    soup = BeautifulSoup(r.text, "lxml")
    
    all_tags = soup.find_all("div", class_="tags")
    
    for a in all_tags:
        for c in a.find_all(recursive=False):
            if "life" in c.get_text().lower():
                parent = a.find_parent("div")
                if parent:
                    quote_tag = parent.select_one("span.text")
                    author_tag = parent.select_one("small.author")
                    if quote_tag and author_tag:
                        ans.append([quote_tag.get_text(strip=True),
                                    author_tag.get_text(strip=True)])
    
    next_page = soup.select_one("li.next > a")
    if next_page:
        url = next_page["href"]
    else:
        url = None

df = pd.DataFrame(ans, columns=["quote", "author"])
df
```

```python
page_count=1
while True:
    URL=f"https://quotes.toscrape.com/page/{page_count}/"
    res=requests.get(URL)
    soup=BeautifulSoup(res.text,"lxml")
    quotes=soup.select("div.quote")
    if not quotes:
        print("no valid pages anymore..")
        break
    with open (f"new/quotes{page_count}.html","w",encoding="utf-8")as f:
         f.write(res.text)
         print(f"download data  from page{page_count}")
    page_count=page_count+1
```

```python
with open("new/quotes1.html", "r", encoding="utf-8") as f:
    htmlContent=f.read()
soup=BeautifulSoup(htmlContent,"lxml")
life_quotes=[]
all_quotes=soup.select("div.quote")
for q in all_quotes:
    all_tags=[]
    for tag in q.select(".tags .tag"):
        all_tags.append(tag.get_text())
    if "life" in all_tags:
        text=q.select_one("span.text").get_text()
        author=q.select_one("small.author").get_text()
    
        life_quotes.append([text,author]) 

```
