```python
#create
arr1=np.zeros((2,3),dtype="int64")
print(arr1)
arr2=np.ones((2,3),dtype="int64")
print(arr2)

arr3=np.eye((3),dtype="int64")
print(arr3)
arr4=np.full((3,4),100)
print(arr4)

arr5=np.arange(0,10,2)  #  start 0  and  print  0 2 4 .....8  
arr6=np.arange(1,10,2)  #  start 0  and  print  1 3 5 .....9
print(arr5)
print(arr6)
arr6=np.linspace(1,100,3)#evenly_space_arrays
print(arr6)
```

```python
arr=np.array([[1,2,3],[3,4,5],[1,2,3]])
print(arr.shape)
print(arr.size)
print(arr.dtype)
print(arr.ndim)

float_arr=arr.astype(np.float64)#  change  data_type
print(float_arr.dtype)
```

```python
#  important  operation  on  numpy_arrays
arr=np.array([[1,2,3],[4,5,6]])
reshaped=arr.reshape((3,2))
print(reshaped)
flattened=arr.flatten()  # 2d to  1d   array
print(flattened)
```

```python
#indexing an  array
arr=np.array([[1,2,3],[4,5,6]])
print(arr[:,0].sum())
```

```python
#fancy _indexing
arr=np.array([1,2,3,4,5]);
idx=[1,3,4]
print(arr[idx])
print(arr[arr%2==1])
```

```python
#slicing
arr=np.array([1,2,3,4,5])
print(arr[1:4])
print(arr[1:])
print(arr[:4])
print(arr[::2])
```

```python
#3d  arrays

arr=np.array([[
  [1,2,3],
  [4,5,6]
],[
  [1,2,3],
  [4,5,6]
]
]       
 )
print(np.sum(arr,axis=0))  #  like   simple   matrix  sum   
print(np.sum(arr,axis=1))  # individual  matrix row sum 
print(np.sum(arr,axis=2))  # individual  matrix column sum 
print(arr[0,8:10,1:3])
arr[:,0,:]=99  #  first  row  of  both layers  and  replace

```

```python
#normalize
arr=np.array([[1,2],[3,4]])
mean=np.mean(arr)
std=np.std(arr)
normalize_arr=(arr-mean)/std
print(normalize_arr)
```

```python
#mathmatic_function

arr=np.array([1,2,4,5,7])
print(np.min(arr))
print(np.max(arr))
print(np.armin(arr)) # return  index
print(np.armax(arr))   # return  index
print(np.std(arr))   # return  index
print(np.mean(arr))   # return  index
print(np.median(arr))   # return  index
print(np.var(arr))   # return  index
print(np.square(arr))
print(np.sqrt(arr))
print(np.pow(arr,3))
print(np.log10(arr))
print(np.round(arr))
print(np.floor(arr))
print(np.ceil(arr))
print(np.trunc(arr))  # only  integer  part  return
print(np.unique(arr)) 
print(np.sort(arr))  
print(np.sort(arr)) 
print(np.abs(arr)) 

```
